```kotlin
val text = File("dayFive.txt").readText()
val blocks = text.split("\n\n")

val seeds = Regex("\\d+").findAll(blocks[0])
    .map { it.value.toLong() }
    .toMutableList()

val pairRangeMap = mutableMapOf<String, List<Pair<LongRange, LongRange>>>()

for (index in 1 until blocks.size) {

    val destinationRangeStarts = mutableListOf<Long>()
    val sourceRangeStarts = mutableListOf<Long>()
    val rangeLengths = mutableListOf<Long>()

    val title = blocks[index].split("\n")[0]

    Regex("\\d+").findAll(blocks[index])
        .map { it.value.toLong() }
        .forEachIndexed { valIndex, value ->
            if (valIndex % 3 == 0) {
                destinationRangeStarts.add(value)
            } else if (valIndex % 3 == 1) {
                sourceRangeStarts.add(value)
            } else if (valIndex % 3 == 2) {
                rangeLengths.add(value)
            }
        }

    pairRangeMap.put(title,
        destinationRangeStarts.zip(sourceRangeStarts).zip(rangeLengths)
            .map {
                LongRange(it.first.second, (it.first.second + it.second) - 1) to LongRange(
                    it.first.first, (it.first.first
                            + it.second) - 1
                )
            }
    )
}


for ((seedIndex, seed) in seeds.withIndex()) {
    var tempSeed = seed
    for (entry in pairRangeMap) {
        val pairRange = entry.value.firstOrNull { tempSeed in it.first } ?: continue
        val sourceValue = pairRange.second.first + (tempSeed - pairRange.first.first).absoluteValue
        tempSeed = sourceValue
        seeds.set(seedIndex, sourceValue)
    }
}

seeds.min()
```

```kotlin
import java.io.File

val text = File("dayFive.txt").readText()
var blocks = text.split("\n\n")
var min = Long.MAX_VALUE

val sectionToExtract = blocks[0]

blocks = blocks.reversed().dropLast(1)

val permutations = Regex("\\d+").findAll(sectionToExtract)
    .map { it.value.toLong() }
    .chunked(2) { (first, second) ->
        first..first + second - 1
    }.sortedBy {
        it.last
    }.toList()

val pairRangeMap = mutableMapOf<String, List<Pair<LongRange, LongRange>>>()

for (index in 1 until blocks.size) {

    val destinationRangeStarts = mutableListOf<Long>()
    val sourceRangeStarts = mutableListOf<Long>()
    val rangeLengths = mutableListOf<Long>()

    val title = blocks[index].split("\n")[0]

    Regex("\\d+").findAll(blocks[index])
        .map { it.value.toLong() }
        .forEachIndexed { valIndex, value ->
            if (valIndex % 3 == 0) {
                destinationRangeStarts.add(value)
            } else if (valIndex % 3 == 1) {
                sourceRangeStarts.add(value)
            } else if (valIndex % 3 == 2) {
                rangeLengths.add(value)
            }
        }

    pairRangeMap.put(title,
        sourceRangeStarts.zip(destinationRangeStarts).zip(rangeLengths)
            .map {
                LongRange(it.first.second, (it.first.second + it.second) - 1) to LongRange(
                    it.first.first, (it.first.first
                            + it.second) - 1
                )
            }.reversed()
    )
}

for (source in 1..permutations.last().last) {
    var seed = source.toLong()
    for (entry in pairRangeMap) {
        val pairRange = entry.value.firstOrNull { seed in it.first } ?: continue
        val sourceValue = pairRange.second.first + (seed - pairRange.first.first).absoluteValue
        seed = sourceValue
    }
    
    println(source)

    if (permutations.any { it.contains(seed) }) {
        println(seed)
        println(source)
        min = source
        break
    }
}

min
```
